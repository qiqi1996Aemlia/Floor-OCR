```python
import cv2
import numpy as np
from pdf2image import convert_from_path
import matplotlib.pyplot as plt

def extract_green_from_floorplan(pdf_path, dpi=300):
    image = convert_from_path(pdf_path, dpi)[0]
    img_rgb = np.array(image)  # PIL → NumPy
    img_bgr = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR)

    # 转换为 HSV
    hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)

    # 定义绿色范围 (HSV)
    lower_green = np.array([35, 40, 40])   # 根据实际颜色调整
    upper_green = np.array([85, 255, 255])

    # 掩膜
    mask = cv2.inRange(hsv, lower_green, upper_green)
    result = cv2.bitwise_and(img_bgr, img_bgr, mask=mask)

    return img_bgr, mask, result
```

```python
import cv2
import numpy as np
from PIL import Image

def enhance_table_lines(pil_img, kernel_size=3, iterations=1):
    # 转换为 NumPy 格式
    img_np = np.array(pil_img.convert("L"))  # 灰度图

    # 二值化（自动选择阈值）
    _, binary = cv2.threshold(img_np, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # 膨胀操作
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (kernel_size, kernel_size))
    dilated = cv2.dilate(binary, kernel, iterations=iterations)

    # 反转回来（让线条变黑）
    result = cv2.bitwise_not(dilated)

    # 转为 PIL 图像
    return Image.fromarray(result)
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from PIL import Image
import numpy as np

def visualize_and_crop_table_cells(image, table_result, draw_text=False):
    """
    image: PIL.Image
    table_result: TableResult or list of TableResult
    return: cell_images -> dict[(row_id, col_id)] = PIL.Image.crop(...)
    """
    if isinstance(table_result, list):
        table_result = table_result[0]

    fig, ax = plt.subplots(1, figsize=(16, 12))
    ax.imshow(image)

    cell_images = {}
    img_np = np.array(image)

    for cell in table_result.cells:
        x1, y1, x2, y2 = map(int, cell.bbox)  # ensure integers for slicing
        rect = Rectangle((x1, y1), x2 - x1, y2 - y1,
                         linewidth=2,
                         edgecolor='red' if cell.is_header else 'blue',
                         facecolor='none')
        ax.add_patch(rect)

        if draw_text:
            ax.text(x1 + 5, y1 + 5, f"R{cell.row_id},C{cell.col_id}",
                    fontsize=8, color='black',
                    bbox=dict(facecolor='white', alpha=0.5, edgecolor='none'))

        # Crop and store each cell
        cropped_img = image.crop((x1, y1, x2, y2))  # (left, upper, right, lower)
        cell_images[(cell.row_id, cell.col_id)] = cropped_img

    ax.axis('off')
    plt.show()

    return cell_images
```

```python
def remove_large_boxes_from_mask(image, mask, area_threshold_ratio=0.8,
                                  margin_ratio_w=0.07, margin_ratio_h=0.13):
    h, w = mask.shape
    image_area = h * w

    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
    closed = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    dilated = cv2.dilate(closed, kernel, iterations=1)

    contours, _ = cv2.findContours(dilated, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    cropped_img = image
    cropped_mask = mask

    for cnt in contours:
        x, y, box_w, box_h = cv2.boundingRect(cnt)
        box_area = box_w * box_h
        percent = box_area / image_area

        if percent > area_threshold_ratio:
            kw = int(box_w * margin_ratio_w)
            kh = int(box_h * margin_ratio_h)

            # 裁剪区域，防止越界
            x1 = max(x + kw, 0)
            y1 = max(y + kh, 0)
            x2 = min(x + box_w - kw, w)
            y2 = min(y + box_h - kh, h)

            cropped_img = image[y1:y2, x1:x2]
            cropped_mask = mask[y1:y2, x1:x2]
            break  # 只处理第一个大框即可

    return cropped_img, cropped_mask
```

```python
def merge_mask_and_segment(img_bgr, mask, min_area=100, max_area = 600, draw_on_original=True, mode='multi'):
    """
    参数：
    - img_bgr: 原始图像
    - mask: 二值图像（目标区域为白，背景为黑）
    - min_area: 忽略小目标的阈值
    - draw_on_original: 是否在原图上绘制框
    - mode: 'multi' 或 'single'；single 表示只提取一个目标（最大或合并后区域）
    """
    vis_img = img_bgr.copy() if draw_on_original else None
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
    closed = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    dilated = cv2.dilate(closed, kernel, iterations=1)

    contours, _ = cv2.findContours(dilated, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    templates = []

    if mode == 'single':
        # 合并所有轮廓为一个最大框
        all_x, all_y, all_x2, all_y2 = [], [], [], []
        for cnt in contours:
            x, y, w, h = cv2.boundingRect(cnt)
            area = w * h
            aspect_ratio = w / h if h != 0 else 0
            if area < min_area or area > max_area or aspect_ratio > 5 or aspect_ratio < 0.2:
                continue
            all_x.append(x)
            all_y.append(y)
            all_x2.append(x + w)
            all_y2.append(y + h)

        if all_x:  # 有有效轮廓
            x1, y1 = min(all_x), min(all_y)
            x2, y2 = max(all_x2), max(all_y2)
            w, h = x2 - x1, y2 - y1

            template_img = img_bgr[y1:y2, x1:x2]
            templates.append({
                "id": "E01",
                "box": (x1, y1, w, h),
                "template": template_img
            })

            if draw_on_original:
                cv2.rectangle(vis_img, (x1, y1), (x2, y2), (0, 0, 255), 2)
                cv2.putText(vis_img, "E01", (x2 + 5, y1 + 15),
                            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), thickness=4)
    else:
        for idx, cnt in enumerate(contours):
            x, y, w, h = cv2.boundingRect(cnt)
            area = w * h
            aspect_ratio = w / h if h != 0 else 0

            if area < min_area or area > max_area or aspect_ratio > 5 or aspect_ratio < 0.2:
                continue

            template_img = img_bgr[y:y+h, x:x+w]
            templates.append({
                "id": f"E{idx+1:02d}",
                "box": (x, y, w, h),
                "template": template_img
            })

            if draw_on_original:
                cv2.rectangle(vis_img, (x, y), (x+w, y+h), (0, 0, 255), 2)
                cv2.putText(vis_img, f"E{idx+1:02d}", (x + w + 5, y + 15),
                            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), thickness=4)

    print(f"✅ {'合并' if mode=='single' else '正常'}后分割出 {len(templates)} 个图标模板")

    if draw_on_original and vis_img is not None:
        plt.figure(figsize=(14, 12))
        plt.imshow(cv2.cvtColor(vis_img, cv2.COLOR_BGR2RGB))
        plt.title("Detected Templates on Original Image")
        plt.axis('off')
        plt.show()

    return templates, vis_img if draw_on_original else None
```

```python
from skimage.metrics import structural_similarity as ssim
import cv2
import numpy as np

def rotate_image(image, angle):
    """旋转图像，保边缘，避免黑框填充失真"""
    h, w = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    return cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_NEAREST, borderValue=255)

def resize_with_padding(img, target_size=(100, 100)):
    """保持纵横比缩放并居中填白"""
    h, w = img.shape[:2]
    scale = min(target_size[0] / h, target_size[1] / w)
    new_h, new_w = int(h * scale), int(w * scale)
    resized = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_NEAREST)

    padded = np.full(target_size, 255, dtype=np.uint8)
    pad_y = (target_size[0] - new_h) // 2
    pad_x = (target_size[1] - new_w) // 2
    padded[pad_y:pad_y+new_h, pad_x:pad_x+new_w] = resized
    return padded

def match_patch_ssim_rotation(patch_img, template_list, size=(100, 100), angles=[0, 90, 180, 270], score_threshold=0.5):
    """匹配 patch 和模板列表，考虑旋转，返回最优匹配的 ID 和名字"""
    patch_gray = cv2.cvtColor(patch_img, cv2.COLOR_BGR2GRAY)
    patch_gray = resize_with_padding(patch_gray, size)

    best_id = None
    best_name = None
    best_score = -1

    for t in template_list:
        template = t["img"]
        template_gray = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
        template_gray = resize_with_padding(template_gray, size)

        for angle in angles:
            rotated_template = rotate_image(template_gray, angle)
            score = ssim(patch_gray, rotated_template)

            if score > best_score:
                best_score = score
                best_id = t["id"]
                best_name = t["name"]

    if best_score < score_threshold:
        return None, None, best_score

    return best_id, best_name, best_score

```

```python
import torch
import clip
from PIL import Image
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# 1. 加载模型
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-L/14", device=device)

from PIL import Image
import cv2
import numpy as np


# 2. 提取图像向量
def get_image_embedding(pil_img):
    img_pre = preprocess(pil_img).unsqueeze(0).to(device)
    with torch.no_grad():
        embedding = model.encode_image(img_pre)
    return embedding / embedding.norm(dim=-1, keepdim=True)

# 3. 计算匹配
def match_patch_clip(patch_pil, symbol_templates):
    patch_embed = get_image_embedding(patch_pil)

    best_score = -1
    best_id, best_name = None, None

    for symbol in symbol_templates:
        template_embed = get_image_embedding(Image.fromarray(symbol['img']))
        score = cosine_similarity(patch_embed.cpu().numpy(), template_embed.cpu().numpy())[0][0]
        if score > best_score:
            best_score = score
            best_id = symbol['id']
            best_name = symbol['name']

    return best_id, best_name, best_score
```

```python
import torch
import timm
import numpy as np
import cv2
from PIL import Image
import torchvision.transforms as transforms
from sklearn.metrics.pairwise import cosine_similarity

# ✅ 1. 模型加载（可以换成你要的精度高的模型）
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model_name = 'vit_base_patch16_224'
model = timm.create_model(model_name, pretrained=True, num_classes=0).to(device)
model.eval()

# ✅ 2. 图像预处理（和模型输入一致）
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=(0.5, 0.5, 0.5), std=(0.5, 0.5, 0.5)
    )
])

# ✅ 3. 提取图像特征向量
def get_image_embedding_timm(pil_img):
    img_tensor = transform(pil_img).unsqueeze(0).to(device)
    with torch.no_grad():
        embedding = model(img_tensor)
    return embedding / embedding.norm(dim=-1, keepdim=True)

# ✅ 4. 匹配 patch 与 symbol template
def match_patch_timm(patch_pil, symbol_templates):
    patch_embed = get_image_embedding_timm(patch_pil)

    best_score = -1
    best_id, best_name = None, None

    for symbol in symbol_templates:
        template_img = Image.fromarray(symbol['img']).convert("RGB")
        template_embed = get_image_embedding_timm(template_img)

        score = cosine_similarity(patch_embed.cpu().numpy(), template_embed.cpu().numpy())[0][0]

        if score > best_score:
            best_score = score
            best_id = symbol['id']
            best_name = symbol['name']

    return best_id, best_name, best_score
```
